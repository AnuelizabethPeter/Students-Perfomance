# 2. What do you think about the students' parental level of education?

```python
 counts=df['parental level of education'].value_counts().to_dict()  
```

```python
import seaborn as sns
import  matplotlib.pyplot as plt
plt.title("Parental level of Education", bbox={'facecolor':'.9', 'pad':5})

plt.pie([float(counts[v]) for v in counts], labels=[str(k) for k in counts], radius=1.2,autopct='%1.1f%%')
plt.show()
```

```python
FINDINGS:
    From the above chart it is evident that majority of parents either they went to college(22.2%) or they holds an associate
    degree(22.6%).Almost 20% of parents are high school graduates also,18% of parents were admitted to high school.
    11.8% of parents are under graduates,But only few ,that is 5.9% of parents, are post graduates.
```

# 3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

```python
df.groupby(['gender',"test preparation course"]).describe().T
```

# MATH TEST SCORE

```python
plotdata = pd.DataFrame({

    "Test prepration course :completed":[72.988506,67.108696],

    "Test prepration course : not completed":[66.467532,61.320359]}, index=["Male", "Female"])

plotdata.plot(kind="bar",figsize=(6, 7))

plt.title("Students Perfomance : Gender and Test Preparation Course")

plt.xlabel("")

plt.ylabel("Average Marks in Maths")
```

```python
FINDINGS:
    Male students scored more marks in math test than that of female students.
    Average mark of male students who attented the test preparation course is 72.98 and that of male students who doesn't 
    attented the course is 66.46. It clearly shows test preparation course was beneficial to attain more marks.
    Average mark of male students who doesnt attented the test preparation course   and female students who 
    attented the course is almost same ,it is 66.46 and 67.10 respectively.
    Like male students ,average mark of female students who doesnt attented the test preparation course is less than that 
    who attented the course,the mean averages are 61.32 and 67.10 respectively.
```
